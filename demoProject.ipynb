```python
sensational_phrases = [
    "shocking", "exposed", "unbelievable", "must see",
    "revealed", "disturbing", "bombshell", "explosive"
]

clickbait_phrases = [
    "you won't believe", "what happens next",
    "they don't want you to know",
    "secret", "hidden truth", "this will shock you"
]

hyperbole_words = [
    "always", "never", "everyone", "nobody",
    "guaranteed", "proven", "completely",
    "absolutely", "everything", "forever"
]
```

```python
def sensational_density(text):
    t = text.lower()
    count = sum(t.count(p) for p in sensational_phrases)
    words = len(t.split())
    return count / words if words else 0
```

```python
def clickbait_density(text):
    t = text.lower()
    count = sum(t.count(p) for p in clickbait_phrases)
    words = len(t.split())
    return count / words if words else 0
```

```python
def hyperbole_density(text):
    words_list = text.lower().split()
    count = sum(w in hyperbole_words for w in words_list)
    words = len(words_list)
    return count / words if words else 0
```

```python
import re

def caps_punct_intensity(text):
    caps_words = re.findall(r"\b[A-Z]{2,}\b", text)
    exclam = text.count("!")
    questions = text.count("?")
    words = len(text.split())
    score = len(caps_words) + exclam + questions
    return score / words if words else 0
```

```python
from textblob import TextBlob

def emotional_intensity(text):
    polarity = TextBlob(text).sentiment.polarity
    return abs(polarity)
```

```python
def matrix1_features(text):
    return {
        "sensational_density": sensational_density(text),
        "clickbait_density": clickbait_density(text),
        "hyperbole_density": hyperbole_density(text),
        "caps_punct_intensity": caps_punct_intensity(text),
        "emotional_intensity": emotional_intensity(text)
    }
```

```python
import requests
from bs4 import BeautifulSoup
from newspaper import Article
def is_valid_article(text):
    sentences = [s for s in text.split(".") if len(s.strip()) > 40]
    return len(sentences) >= 5

def extract_article(url):
    # Try Newspaper3k
    try:
        article = Article(url, browser_user_agent="Mozilla/5.0")
        article.download()
        article.parse()

        if article.text and len(article.text) > 200:
            if not is_valid_article(article.text):
                return {"url": url, "error": "Not a single article page"}

            return {
                "url": url,
                "title": article.title,
                "text": article.text
            }
    except:
        pass

    # Fallback BeautifulSoup
    try:
        headers = {"User-Agent": "Mozilla/5.0"}
        html = requests.get(url, headers=headers, timeout=10).text
        soup = BeautifulSoup(html, "html.parser")

        paragraphs = soup.find_all("p")
        text = " ".join(p.get_text(strip=True) for p in paragraphs)

        if len(text) < 200 or not is_valid_article(text):
            return {"url": url, "error": "Not a single article page"}

        return {
            "url": url,
            "title": soup.title.string if soup.title else "",
            "text": text
        }

    except Exception as e:
        return {"url": url, "error": str(e)}
```

```python
import pandas as pd

def extract_from_urls(url_list):
    records = []
    for u in url_list:
        art = extract_article(u)
        records.append(art)
    return pd.DataFrame(records)
```

```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np

# load embedding model once
embedder = SentenceTransformer("all-MiniLM-L6-v2")


def headline_body_similarity(title, text):
    """
    Semantic similarity between headline and article body
    Range: 0–1
    """
    if not title or not text:
        return 0.0

    emb_title = embedder.encode([title])
    emb_body = embedder.encode([text[:1000]])  # limit length

    sim = cosine_similarity(emb_title, emb_body)[0][0]
    return float(sim)


def topic_drift(text):
    sentences = [s.strip() for s in text.split(".") if len(s.strip()) > 20]

    if len(sentences) < 2:
        return 0.0

    embs = embedder.encode(sentences)
    sims = []

    for i in range(len(embs) - 1):
        s = cosine_similarity([embs[i]], [embs[i+1]])[0][0]
        sims.append(s)

    return float(np.var(sims))

def matrix2_features(title, text):
    return {
        "headline_body_similarity": headline_body_similarity(title, text),
        "topic_drift": topic_drift(text)
    }
```

```python
import tldextract

def extract_domain(url):
    ext = tldextract.extract(url)
    return ext.domain + "." + ext.suffix
```

```python
source_stats = {
    "bbc.com": {"fake_ratio": 0.05},
    "reuters.com": {"fake_ratio": 0.03},
    "nytimes.com": {"fake_ratio": 0.04},
    "foxnews.com": {"fake_ratio": 0.15},
    "randomblog.com": {"fake_ratio": 0.70},
    "unknownnews.net": {"fake_ratio": 0.80}
}
```

```python
def source_reliability(url):
    domain = extract_domain(url)

    if domain in source_stats:
        fake_ratio = source_stats[domain]["fake_ratio"]
        return 1 - fake_ratio

    # unknown source → neutral prior
    return 0.5
```

```python
def matrix3_features(url):
    return {
        "source_domain": extract_domain(url),
        "source_reliability": source_reliability(url)
    }
```

```python
import requests
from bs4 import BeautifulSoup
from playwright.sync_api import sync_playwright

# =========================
# CONFIG
# =========================
MIN_TEXT_LENGTH = 500

BLOCKED_DOMAINS = [
    "linkedin.com",
    "instagram.com",
    "facebook.com",
    "twitter.com",
    "x.com",
    "docs.google.com",
    "drive.google.com"
]

BLOCK_KEYWORDS = [
    "login",
    "sign in",
    "sign up",
    "captcha",
    "verify you are human",
    "access denied",
    "enable javascript"
]


# =========================
# HELPERS
# =========================
def is_blocked_domain(url: str) -> bool:
    return any(d in url.lower() for d in BLOCKED_DOMAINS)


def looks_blocked(text: str) -> bool:
    t = text.lower()
    return any(k in t for k in BLOCK_KEYWORDS)


def clean_html(html: str) -> str:
    soup = BeautifulSoup(html, "html.parser")

    for tag in soup(["script", "style", "noscript", "header", "footer", "nav"]):
        tag.extract()

    return soup.get_text(" ", strip=True)


# =========================
# METHOD 1 — REQUESTS
# =========================
def try_requests(url: str):
    try:
        headers = {
            "User-Agent": "Mozilla/5.0",
            "Accept-Language": "en-US,en;q=0.9"
        }

        r = requests.get(url, headers=headers, timeout=10)

        if r.status_code != 200:
            return None

        text = clean_html(r.text)

        if len(text) < MIN_TEXT_LENGTH or looks_blocked(text):
            return None

        return text

    except Exception:
        return None


# =========================
# METHOD 2 — PLAYWRIGHT
# =========================
def try_playwright(url: str):
    try:
        with sync_playwright() as p:
            browser = p.chromium.launch(headless=True)
            page = browser.new_page()

            page.goto(url, timeout=30000)
            page.wait_for_load_state("networkidle")

            text = page.inner_text("body")
            browser.close()

            if len(text) < MIN_TEXT_LENGTH or looks_blocked(text):
                return None

            return text

    except Exception:
        return None


# =========================
# METHOD 3 — READER API
# =========================
def try_reader(url: str):
    try:
        reader_url = f"https://r.jina.ai/{url}"
        r = requests.get(reader_url, timeout=15)

        if r.status_code != 200:
            return None

        text = r.text.strip()

        if len(text) < MIN_TEXT_LENGTH or looks_blocked(text):
            return None

        return text

    except Exception:
        return None


# =========================
# CORE EXTRACTION PIPELINE
# =========================
def extract_url_text(url: str, verbose=True):
    if is_blocked_domain(url):
        return {
            "status": "failed",
            "reason": "Domain blocks automated access",
            "method": None,
            "text": None
        }

    if verbose:
        print("Trying requests...")
    text = try_requests(url)
    if text:
        if verbose:
            print("Extracted via requests:", len(text))
        return {"status": "success", "method": "requests", "text": text}

    if verbose:
        print("Trying browser...")
    text = try_playwright(url)
    if text:
        if verbose:
            print("Extracted via Playwright:", len(text))
        return {"status": "success", "method": "playwright", "text": text}

    if verbose:
        print("Trying reader API...")
    text = try_reader(url)
    if text:
        if verbose:
            print("Extracted via Reader:", len(text))
        return {"status": "success", "method": "reader", "text": text}

    if verbose:
        print("Extraction failed")

    return {
        "status": "failed",
        "reason": "Content blocked, login required, or not extractable",
        "method": None,
        "text": None
    }


# =========================
# GENAI INGESTION WRAPPER
# =========================
def ingest_url_for_genai(url: str, verbose=True):
    result = extract_url_text(url, verbose=verbose)

    if result["status"] == "success":
        return {
            "status": "ok",
            "source": "url",
            "method": result["method"],
            "text": result["text"]
        }

    return {
        "status": "error",
        "source": "url",
        "message": (
            "This website blocks automated reading or requires login. "
            "Please paste the article text or use a public webpage URL."
        )
    }
```
